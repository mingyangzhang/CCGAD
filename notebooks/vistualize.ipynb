```python
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
```

```python
data = sio.loadmat("../outputs/cora.mat")
```

```python
scores = data['scores'].flatten()
labels = data['labels'].flatten()
pos_probs = data['pos_probs'].flatten()
neg_probs = data['neg_probs'].flatten()
```

```python
fig, (ax0, ax1, ax2) = plt.subplots(3, 1, figsize=(3, 9))
ax0.boxplot([scores[np.where(labels==1)], scores[np.where(labels==0)]], labels=["Abnor", "Nor"])
ax0.tick_params(axis='both', which='major', labelsize=16)
ax0.tick_params(axis='both', which='minor', labelsize=16)
ax0.set_ylabel('Abnormal Score', fontsize=18)
# plt.title('In cluster negative sampling')

ax1.boxplot([pos_probs[np.where(labels==1)], pos_probs[np.where(labels==0)]], labels=["Abnor", "Nor"])
ax1.tick_params(axis='both', which='major', labelsize=16)
ax1.tick_params(axis='both', which='minor', labelsize=16)
ax1.set_ylabel('Pos Probs', fontsize=18)

ax2.boxplot([neg_probs[np.where(labels==1)], neg_probs[np.where(labels==0)]], labels=["Abnor", "Nor"])
ax2.tick_params(axis='both', which='major', labelsize=16)
ax2.tick_params(axis='both', which='minor', labelsize=16)
ax2.set_ylabel('Neg Probs', fontsize=18)
```
